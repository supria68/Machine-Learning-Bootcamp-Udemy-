# K Nearest Neighbors
  
This notebook is dedicated to understand the usage of KNN algorithm for classification problems. Given a synthetic dataset, use KNN to classify whether the new data belongs to 1 or 0. Explore how the k values can affect the performance of predicting model.  
  
Dataset: A company data consisting of 1000 entries, with 10 features and 1 target class.

```python
# Basic imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

### Loading the Dataset

```python
data = pd.read_csv('Data/KNN_Project_Data')
```

```python
data.head()
```

```python
data.info()
```

### Exploratory Data Analysis

Since this data is artificial, let's use the pairplot to understand the behaviour of each features.

```python
sn.pairplot(data, hue = 'TARGET CLASS', palette = 'viridis')
```

The dataset is completely non-linear. It's obvious that we cannot use logistic regression to classify things. Since the number of entries are also minimal, we can try out the KNN approach to classify data.

### Standardize the Variables

Since there isn't any information on ranges that each feature can take, let's standardize them.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() #object instantiation
```

```python
scaler.fit(data.drop('TARGET CLASS', axis = 1)) #Standardization should be carried out only on the features
```

```python
scaled_features = scaler.transform(data.drop('TARGET CLASS', axis = 1))
```

```python
X = pd.DataFrame(scaled_features, columns = data.columns[:-1])
```

```python
X.head() # feature table
```

```python
y = data['TARGET CLASS'] # target
```

### Using KNN with K = 1

Assume the value of K = 1. Let's train the model on training data and evaluate it on test data. 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors = 1)
knn_model.fit(X_train,y_train)
```

### Prediction and Evaluation

Let's test our classifier on the test set and evaluate the precision, recall, accuracy!

```python
prediction_k1 = knn_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(prediction_k1, y_test))
print('\n')
print(classification_report(prediction_k1,y_test))
```

So, our KNN classifier (k = 1) has an accuracy of 69%. This is pretty bad! 

### Choosing a K Value

Let's choose the value of K for which the cross validation error rate is low or accuracy is high!

```python
from sklearn.model_selection import cross_val_score   
```

```python
error_rate = []
accuracy_rate = []
for k in range(1,40):
    knn_model = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_model,X , y, cv = 10) # Use cross validation sets to choose k when error is low!
    accuracy_rate.append(np.mean(score))
    error_rate.append(1-np.mean(score))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,40), error_rate, color = 'blue', ls = '--', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Cross validation Error rate Vs K')
plt.xlabel('K values')
plt.ylabel('Error rate')
```

For K = 18, the error_rate is lowest! Also for K > 16, the error rate tends to hover around 0.18.  
Let's check what is the accuracy for K > 16 cases.

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,40), accuracy_rate, color = 'green', ls = '--', marker = 'o', markerfacecolor = 'yellow', markersize = 10)
plt.title('Accuracy rate Vs K')
plt.xlabel('K values')
plt.ylabel('Accuracy rate')
```

It's pretty much the same result as the error_rate vs K plot!
In general, either error_rate or accuracy_rate could be considered while choose the value of K.

### KNN Classifier with new K

let's retrain the model with new K = 18, evaluate the accuracy, precision and recall!

```python
knn_model = KNeighborsClassifier(n_neighbors = 18)
knn_model.fit(X_train,y_train)
```

```python
# Evaluation
predictions = knn_model.predict(X_test)
```

```python
print(confusion_matrix(predictions, y_test))
print('\n')
print(classification_report(predictions,y_test))
```

Classifier accuracy is 81%. With K = 18, accuracy is much improved!
